```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def jump(fleas):
    idx = np.random.randint(len(fleas))
    fleas[idx] = (fleas[idx] + 1)%2

N = 400
fleas = np.zeros(N)
sums = []
iter = 3000
for i in range(iter):
    jump(fleas)
    sums.append(np.sum(fleas))
```

```python
Hvals = np.array([s * np.log(s) + (N-s) * np.log(N-s) for s in sums])

plt.figure(figsize=(12,3.3))
plt.subplot(1,2,1)
plt.plot(sums)
plt.xlabel('Time step', fontsize=14)
plt.ylabel('Fleas on dog 1, $s$', fontsize=14)
plt.axhline(N/2, color='gray', linestyle='--')
plt.subplot(1,2,2)
plt.plot(Hvals)
plt.xlabel('Time step', fontsize=14)
plt.ylabel('H', fontsize=14)
```
