```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def jump(fleas):
    idx = np.random.randint(len(fleas))
    fleas[idx] = (fleas[idx] + 1)%2
```

```python
N = 400
fleas = np.zeros(N)
fleaCountOn1 = []
timeSteps = 3000

for i in range(timeSteps):
    jump(fleas)
    fleaCountOn1.append(np.sum(fleas))
    
fleaCountOn1 = np.array(fleaCountOn1)

# Calculates the H value for each time step
Hvals = fleaCountOn1 * np.log(fleaCountOn1) + (N-fleaCountOn1) * np.log(N-fleaCountOn1)

plt.figure(figsize=(12,3.3))
plt.subplot(1,2,1)
plt.plot(fleaCountOn1)
plt.xlabel('Time step', fontsize=14)
plt.ylabel('Fleas on dog 1, $s$', fontsize=14)
plt.axhline(N/2, color='gray', linestyle='--')
plt.subplot(1,2,2)
plt.plot(Hvals)
plt.xlabel('Time step', fontsize=14)
plt.ylabel('H', fontsize=14)
```
